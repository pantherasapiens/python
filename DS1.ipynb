```python
import numpy as np
import matplotlib.pyplot as plt
x = np.array([1,2,3,4])
y = np.array([6,7,8,9])
print(x,y)

plt.plot(x,y)
plt.show
```

```python
z = np.array([[345,45,65],[45,68,87]])
plt.plot(z)
plt.show
```

```python
plt.plot((np.array([8,7,5,3,9])))
plt.show
```

```python
plt.plot(z,marker='*')
```

```python
plt.plot(x,y,'o')
plt.show
```

```python
class PrintPlot:
    def plotting(self,x,y):
        plt.plot(x,y,marker = '*')
        plt.show

p1 = PrintPlot()
p1.plotting(x,y)
```

```python
x = np.array([98,78,65,45,32,12])
y = np.array([57,59,51,53,20,30])

p1.plotting(x,y)
```
